# Open CV cat detection

```python
import cv2
```

```python
faceCascade=cv2.CascadeClassifier(r"C:\Users\user\.conda\envs\dark\reconn\haarcascade_frontalcatface.xml")
```

```python
image=cv2.imread(r"C:\Users\user\.conda\envs\dark\reconn\cats.jpg")
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(image)
```

```python
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

```python
faces=faceCascade.detectMultiScale(
    GR_img,
    scaleFactor=1.0001,
    minNeighbors=5,
    minSize=(50,50)
    
)
print(faces)
```

```python
image=cv2.imread(r"C:\Users\user\.conda\envs\dark\reconn\cats.jpg")
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```
